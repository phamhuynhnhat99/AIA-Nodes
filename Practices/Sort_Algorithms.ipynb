```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
r = lambda: np.random.randint(0, 199999999)
```

```python
def comparison(x, y):
    if x < y:
        return -1
    elif x > y:
        return 1
    else:
        return 0
    

def merge_sort(a, l, r):
    if l >= r:
        return
    m = (l + r) // 2
    merge_sort(a, l, m)
    merge_sort(a, m+1, r)
    b = a[l:r+1].copy()
    il = l
    ir = m+1
    for i in range(r-l+1):
        if il <= m:
            if ir > r or comparison(b[il-l], b[ir-l]) < 0:
                a[l+i] = b[il-l]
                il += 1
            else:
                a[l+i] = b[ir-l]
                ir += 1
        else:
            a[l+i] = b[ir-l]
            ir += 1

            
def quick_sort(a, first_ind, last_ind):
    left = first_ind
    right = last_ind
    pivot = a[np.random.randint(left, right)]
    while left < right:
        while comparison(a[left], pivot) < 0:
            left = left + 1
        while comparison(pivot, a[right]) < 0:
            right = right - 1
        if left <= right:
            a[left], a[right] = a[right], a[left]
            left = left + 1
            right = right - 1
    if first_ind < right:
        quick_sort(a, first_ind, right)
    if left < last_ind:
        quick_sort(a, left, last_ind)
        

def partion(a, l, r):
    pivot = a[r]
    i = l - 1
    for j in range(l, r):
        if comparison(a[j], pivot) <= 0:
            i = i + 1
            a[i], a[j] = a[j], a[i]
    a[i+1], a[r] = a[r], a[i+1]
    return i + 1

    
def quick_sort2(a, l, r):
    if l < r:
        pivot = partion(a, l, r)
        quick_sort2(a, l, pivot-1)
        quick_sort2(a, pivot+1, r)
        
        
def bubble_sort(a, l, r):
    for i in range(l, r):
        for j in range(l, r-i):
            if comparison(a[j], a[j+1]) > 0:
                a[j], a[j+1] = a[j+1], a[j]
        

def selection_sort(a, l, r):
    for i in range(l, r):
        for j in range(i+1, r+1):
            if comparison(a[i], a[j]) > 0:
                a[i], a[j] = a[j], a[i]


def insertion_sort(a, l, r):
    for i in range(l+1, r+1):
        key = a[i]
        j = i-1
        while j >= 0 and a[j] > key:
            a[j+1] = a[j]
            j = j - 1
        a[j+1] = key
```

<h1> Non-Decreasing Array </h1>

```python
arr = [100000-_ for _ in range(100000)]

x = list()
for i in range(100):
    num = (i+1) * 20
    x.append(num)
    
y_bs = list()
y_ss = list()
y_is = list()
y_ms = list()
y_qs = list()
y_qs2 = list()

for num in x:
    
    a_bs = arr[:num].copy()
    start_time = time.time()
    bubble_sort(a_bs, 0, len(a_bs) - 1)
    end_time = time.time()
    y_bs.append(end_time - start_time)
    
    a_ss = arr[:num].copy()
    start_time = time.time()
    selection_sort(a_ss, 0, len(a_ss) - 1)
    end_time = time.time()
    y_ss.append(end_time - start_time)
    
    a_is = arr[:num].copy()
    start_time = time.time()
    insertion_sort(a_is, 0, len(a_is) - 1)
    end_time = time.time()
    y_is.append(end_time - start_time)
    
    a_ms = arr[:num].copy()
    start_time = time.time()
    merge_sort(a_ms, 0, len(a_ms) - 1)
    end_time = time.time()
    y_ms.append(end_time - start_time)
    
    a_qs = arr[:num].copy()
    start_time = time.time()
    quick_sort(a_qs, 0, len(a_qs) - 1)
    end_time = time.time()
    y_qs.append(end_time - start_time)
    
    a_qs2 = arr[:num].copy()
    start_time = time.time()
    quick_sort2(a_qs2, 0, len(a_qs2) - 1)
    end_time = time.time()
    y_qs2.append(end_time - start_time)

plt.figure(figsize=(16,10))
plt.plot(x, y_bs, label = "Bubble Sort")
plt.plot(x, y_is, label = "Insertion Sort")
plt.plot(x, y_qs2, label = "Quick Sort 2")
plt.plot(x, y_qs, label = "Quick Sort")
plt.plot(x, y_ms, label = "Merge Sort")
plt.legend()
plt.show()
```
